```python
# Let's check the percentage of missing values in each column of the dataframe
# Columns with high number of missing values will be ommited from the study
#Let's create a function to check the %missing values

def missingdataincol(data):
    total = data.isnull().sum().sort_values(ascending = False)# Sum of missing values in a column
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)#Percent missing data
    return pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])#Concatenated df of count and percent
```

```python
# Let's create new column for average ratings of the customer
# Replacing NaN by 0 will give more weightage to rating with less value. Thus, creating a low overall rating.
# The above method gives due congnisence to high risk attributed to an entry by a rating by giving it low rating.
df_target1_useful1['Average Ratings']=(df_target1_useful1['EXT_SOURCE_1']+df_target1_useful1['EXT_SOURCE_2']+df_target1_useful1['EXT_SOURCE_3'])/3
df_target1_useful1.head(10)
```

# 4.Univariate analysis of number of employment days(Days_Employed) of defaulters:

```python
df_e=pd.DataFrame(df_target1_final['DAYS_EMPLOYED'].value_counts())
```

```python
df_e
```

```python
# From the above DataFrame we can say that, entry with "days_employed=365243" is an outlier.
#Negative value of days shows since number of days back from today.
```

```python
def days(x): # Function to remove outlier and convert negative number of days to positive value
    if x<0:
        return(x*-1)
    else:
        del(x)
def days1(x):
    if x<0:
        return(x*-1)
    else:
        return(x)

df_d=df_target1_final['DAYS_EMPLOYED'].apply(days) #apply function days on target column pensioners excluded.
df_d=df_d.dropna() #This will delete rows which had outlier as entries i.e pensioners excluded.

plt.figure(figsize=(8,15))
 
plt.subplot(211)
sns.distplot(df_d,rug=True,hist=False)## Plot excluding pensioners.

plt.subplot(212)
df_dpen=df_target1_final['DAYS_EMPLOYED'].apply(days1)## Plot including pensioners.
sns.distplot(df_dpen,rug=True,hist=False,color='g')
```

# From the above univariate analysis of Number of employment days('DAYS_EMPLOYED') of defaulters:

> Mostly defaulters have less than 8000 number of employmnet days.

> Those applicant with more than 10000 number of employment days are less likely to default on payment.

> Those with more than 15000 and less than 30000 days of employment are least likely to default on payment.

> Those with more than 30000 days of employment(pensioners) are likely to default on payment.
