```python
def miss(x):
    total = x.isnull().sum().sort_values(ascending = False)
    percent = (x.isnull().sum()/x.isnull().count()*100).sort_values(ascending = False)
    df=pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return(df)
```

```python
df_miss=miss(df4)
df_miss.head(2)
```

```python
# df_miss is DataFrame which shows the entries with missing value percentage less than 40%


df_miss=df_miss.loc[df_miss['Percent']<40]
df_miss
```
